**Question:**

1. *Use PyMC to provide Bayesian inference for the paramaters associated with a sample of normal data where your prior for theta is a normal distribution and your prior for $\tau$ is a gamma distribution. Provide diagnostic assessments of the performance of your algorithm.*

2. *Use PyMC to provide Bayesian inference for the paramaters associated with a sample of normal data where your prior for theta is a non normal distribution and your prior for $\tau$ is a non-gamma distribution. Provide diagnostic assessments of the performance of your algorithm.*

3. *Use PyMC to provide Bayesian inference for the paramaters associated with a sample of normal data where your prior for theta is a yet another different again non normal distribution and your prior for $\tau$ is a yet another different again non-gamma distribution. Provide diagnostic assessments of the performance of your algorithm.*

**Answer:**

**2.1:**

```python
import pymc as pm
import numpy as np
import arviz as az
import matplotlib.pyplot as plt

# Generate sample data from Normal distribution
np.random.seed(42)
n = 100
true_theta, true_tau = 0, 1
data = np.random.normal(loc=true_theta, scale=1/np.sqrt(true_tau), size=n)

# Model 1: Normal Prior for theta, Gamma Prior for tau

with pm.Model() as model_1:
    # Define priors
    theta = pm.Normal("theta", mu=0, sigma=1)  # Normal prior
    tau = pm.Gamma("tau", alpha=2, beta=2)  # Gamma prior

    # Likelihood
    x_obs = pm.Normal("x_obs", mu=theta, tau=tau, observed=data)

    # Sampling
    trace_1 = pm.sample(draws=5000, chains=2, tune=1000)

# Posterior Analysis
print("### Model 1: Normal-Gamma ###")
print(az.summary(trace_1, hdi_prob=0.95))

# Trace plots
az.plot_trace(trace_1)
plt.show()
```

```python
# Credible interval for theta & tau

print("95% Credible Interval for theta:", az.hdi(trace_1.posterior["theta"], hdi_prob=0.95))
print("95% Credible Interval for tau:", az.hdi(trace_1.posterior["tau"], hdi_prob=0.95))
```

The diagnostic assessment of Model 1 (Normal-Gamma) indicates that the MCMC chains have mixed well, as seen in the trace plots, which show no visible trends or drifts. The R-hat values for both $\theta$ and $\tau$ are exactly 1.0, confirming that the chains have fully converged and mixed well. Additionally, the effective sample size (ESS) is well above 10,000 for both parameters, indicating highly efficient sampling with minimal autocorrelation. The posterior density plots are smooth and unimodal, with $\theta$ following a normal-like distribution and $\tau$ exhibiting the expected gamma shape, suggesting that the priors and likelihood were well-specified. Overall, the diagnostics confirm that the model is stable, has converged properly, and produces reliable posterior estimates.

**2.2：**

```python
# Model 2: Non-Normal Prior for theta, Non-Gamma Prior for tau
with pm.Model() as model_2:
    # Define priors
    theta = pm.StudentT("theta", nu=5, mu=0, sigma=1)  # Student-T prior
    tau = pm.HalfNormal("tau", sigma=1)  # Half-Normal prior

    # Likelihood
    x_obs = pm.Normal("x_obs", mu=theta, tau=tau, observed=data)

    # Sampling
    trace_2 = pm.sample(draws=5000, chains=2, tune=1000)

# Posterior Analysis
print("### Model 2: T-HalfNormal ###")
print(az.summary(trace_2, hdi_prob=0.95))

# Trace plots
az.plot_trace(trace_2)
plt.show()
```

The diagnostic assessment for Model 2 (T-HalfNormal) shows that the MCMC chains have mixed well, as indicated by the trace plots, which display no clear trends or divergence. The R-hat values for both $\theta$ and $\tau$ are exactly 1.0, confirming that he chains are mixed well, thus leading to proper convergence. The posterior density plots are smooth and unimodal, with 
$\theta$ following a Student-T shape and $\tau$ behaving as expected under a Half-Normal prior.

```python
# Model 3: Uniform Prior for theta, Exponential Prior for tau

with pm.Model() as model_3:
    # Define priors
    theta = pm.Uniform("theta", lower=-2, upper=2)  # Uniform prior
    tau = pm.Exponential("tau", lam=1)  # Exponential prior

    # Likelihood
    x_obs = pm.Normal("x_obs", mu=theta, tau=tau, observed=data)

    # Sampling
    trace_3 = pm.sample(draws=5000, chains=2, tune=1000)

# Posterior Analysis
print("### Model 3: Uniform-Exponential ###")
print(az.summary(trace_3, hdi_prob=0.95))

# Trace plots
az.plot_trace(trace_3)
plt.show()
```

The diagnostic assessment for Model 3 (Uniform-Exponential) confirms that the MCMC chains exhibit good mixing, as the trace plots show no signs of drift or non-stationarity. The R-hat values for both $\theta$ and $\tau$ are exactly 1.0, indicating proper convergence of the chains. The effective sample size (ESS) is high, with values exceeding 9,700 for both parameters, ensuring efficient sampling and low autocorrelation. *First explain how the Markov algorithm of slice sampling as given below works.
```python
def slice_f_at_y(f, x, y, x_grid=np.linspace(0,1,51)):
    
    # find interval of grid points where f(x_grid) > y 
    # then extend the enterval so endpoints f(a)<y and f(b)<y 
    x_grid_delta = x_grid[1]-x_grid[0]
    a,b = x_grid[f(x_grid)>y][[0,-1]]+[-x_grid_delta,x_grid_delta]
    # a,b = x_grid[0,-1]  # make the interval all of x_grid
    
    x_ = a + stats.uniform().rvs()*(b-a)    
    if f(x_)>y:
        return x_,1  # in 1 try if f(x_)>y and "x_ is under f" 
    elif x_ < x:  # or if "x_ was above f on the left side of the interval" 
        x_l,x_r = x_,b
    else:
        x_l,x_r = a,x_  # or if "x_ was above f on the right side of the interval"
    return slice_f_at_y_(f, x, y, x_l, x_r, tot = 2)  # try again with a reduced interval

def slice_f_at_y_(f, x, y, x_l=0, x_r=1, tot=1):
    
    x_ = x_l + stats.uniform().rvs()*(x_r-x_l)
    
    if f(x_)>y:
        return x_,tot
    elif x_ < x:
        x_l = x_
    else:
        x_r = x_
    return slice_f_at_y_(f, x, y, x_l, x_r, tot = tot+1)
```

```python
from scipy import stats 

x_grid = np.linspace(0,1,1000)
f = lambda x: stats.beta(1.5,3).pdf(x)
plt.plot(x_grid, f(x_grid))
    
m = 1000
x = np.zeros([m+1,3])
x[:2,0] = 0.25

plot_trace = 10
for t in range(1,m):
    
    x[t,1] = stats.uniform().rvs()*f(x[t,0])
    if t < plot_trace:
        plt.plot([x[t,0]]*2, [x[t-1,1],x[t,1]], 'k')
    
    x[t+1,0],x[t+1,2] = slice_f_at_y(f, x[t,0], x[t,1])
    if t < plot_trace:
        if t==1:
            plt.plot([x[t,0], x[t+1,0]], [x[t,1]]*2, 'k--', label=str(plot_trace)+ " iterations")
        plt.plot([x[t,0], x[t+1,0]], [x[t,1]]*2, 'k--')
    
plt.hist(x[:,0], density=True, label=str(m)+" iterations\n x values only")
plt.legend();
```

**Answer:**

Slice sampling is an MCMC method used to sample from a probability distribution by introducing an auxiliary variable y that helps define a slice—a region where the function value is above y. Instead of proposing a new value and deciding whether to accept or reject (as in Metropolis-Hastings), slice sampling adaptively determines an interval from which the next sample is drawn. The algorithm above first defines the slice, given the current sample $x_t$, a uniform random variable $y$ is drawn such that $y$ is less than or equal to $f(x_t)$, creating a horizontal slice where the function is greater than $y$. Then, the algorithm find the valid interval, where it locates an interval [a,b] around $x_t$ where $f(x)>y$, ensuring that the next sample is rawn from a valid region. Next, a new x is drawn uniformly from [a, b]. If the new sample satisfies $f(x)>y$, it is accepted. If the sampled x falls outside the valid slice, the interval is adjusted, and a new x is drawn within the reduced region. This process is repeated iteratively to generate a sequence of dependent samples that follow the target distribution.

The plot represents the slice sampling process applied to Beta(1.5,3) distributionm which is the target distribution we are sampling from. The black dashed lines illustrate the slice sampling process. The black dashed lines illustrate the slice sampling process. Each horizontal dashed line represents a randomly chosen threshold y, and the vertical lines depict the valid interval [a, b] from which the next sample is drawn. This visualization helps us see how slice sampling dynamically adjusts the sampling region in each iteration.

For Gibbs sampling, the initial value is set at $x_0 = 0.25$, serving as the starting point for the Markov chain. At each iteration:

1. A random uniform threshold y is drawn such that $y \leq f(x_t)$, defining a horizontal slice beneath the target distribution.
2. The valid interval [a, b] is determined, where $f(x) > y$.
3. A new $x_{t+1}$ is drawn uniformly from this interval, ensuring it remains within the valid region.
4. If necessary, the interval is shrunk iteratively until a valid sample is accepted.
5. The process repeats for m = 1000 iterations, generating a sequence of samples that approximate the target distribution.

By replacing the usual Metropolis-Hastings step with slice sampling, the Gibbs sampler can efficiently explore the distribution without requiring a proposal distribution or acceptance-rejection step.
