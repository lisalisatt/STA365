### Q1
**Question:** 

1. $p(\boldsymbol \beta | \textbf{y}, \textbf{X}, \Sigma=\sigma^2 I)$ for the **linear regression model** likelihood proportional to $\exp \left(-{\frac {1}{2}}(\mathbf {y} -\mathbf{X\boldsymbol \beta})^\top \Sigma^{-1}(\mathbf {y} - \mathbf{X\boldsymbol\beta})\right)$ and a $\mathcal{MVN}(\boldsymbol\beta_0, \boldsymbol\Sigma_\beta)$ for $\boldsymbol\beta$

2. The analytical posterior distribution for $\sigma^2$ the **error variance** of a **linear regression model** with **design matrix** $\mathbf{X}$ assuming $\sigma^2$ has an **inverse-gamma** prior distribution with parameters $\alpha^*$ and $\beta^*$ (unrelated to $\boldsymbol \beta$).

**Answer:** 

**Q1.1:**

Likelihood:  
$p(y|X, \beta, \sigma^2) \propto \exp \left(-\frac{1}{2} (y - X\beta)^T \Sigma^{-1} (y - X\beta) \right)$  

Prior for $\beta$:  
$\beta \sim N(\beta_0, \Sigma_{\beta}) \Rightarrow P(\beta) \propto \exp \left(-\frac{1}{2} (\beta - \beta_0)^T \Sigma_{\beta}^{-1} (\beta - \beta_0) \right)$  

$$
P(\beta | y, X, \Sigma = \sigma^2 I) \propto P(y | X, \beta, \Sigma = \sigma^2 I) P(\beta)
$$  

$$
= \exp \left(-\frac{1}{2\sigma^2} (y - X\beta)^T (y - X\beta) \right) \times \exp \left(-\frac{1}{2} (\beta - \beta_0)^T \Sigma_{\beta}^{-1} (\beta - \beta_0) \right)
$$  

**Expand the likelihood term:**

$$
(y - X\beta)^T (y - X\beta) = y^T y - 2 \beta^T X^T y + \beta^T X^T X \beta
$$  

**Expand the prior term:**

$$
(\beta - \beta_0)^T \Sigma_{\beta}^{-1} (\beta - \beta_0) = \beta^T \Sigma_{\beta}^{-1} \beta - 2\beta^T \Sigma_{\beta}^{-1} \beta_0 + \beta_0^T \Sigma_{\beta}^{-1} \beta_0
$$  

Thus,

$$
P(\beta | y, X, \Sigma = \sigma^2 I) \propto \exp \left( -\frac{1}{2\sigma^2} (y^T y - 2\beta^T X^T y + \beta^T X^T X \beta) \right) \times \exp \left( -\frac{1}{2} (\beta^T \Sigma_{\beta}^{-1} \beta - 2\beta^T \Sigma_{\beta}^{-1} \beta_0 + \beta_0^T \Sigma_{\beta}^{-1} \beta_0) \right)
$$  

$$
\propto \exp \left\{ -\frac{1}{2} ( -2\beta^T X^T y + \beta^T X^T X \beta ) - \frac{1}{2} (\beta^T \Sigma_{\beta}^{-1} \beta - 2\beta^T \Sigma_{\beta}^{-1} \beta_0) \right\}
$$  

$$
= \exp \left\{ -\frac{1}{2} \left[ \beta^T \left(\frac{X^T X}{\sigma^2} + \Sigma_{\beta}^{-1} \right) \beta - 2\beta^T \left( \frac{X^T y}{\sigma^2} + \Sigma_{\beta}^{-1} \beta_0 \right) \right] \right\}
$$  

We recognize this in the form:

$$
-\frac{1}{2} \left[ \beta^T A \beta - 2\beta^T b \right],
$$  

where:  
- $A = \frac{X^T X}{\sigma^2} + \Sigma_{\beta}^{-1}$  
- $b = \frac{X^T y}{\sigma^2} + \Sigma_{\beta}^{-1} \beta_0$  

**Completing the square:**

$$
\beta^T A \beta - 2\beta^T b = (\beta - A^{-1} b)^T A (\beta - A^{-1} b) - b^T A^{-1} b
$$  

This means:

$$
\beta | y, X, \sigma^2 \sim N(A^{-1} b, A^{-1}) = N\left( \left(\frac{X^T X}{\sigma^2} + \Sigma_{\beta}^{-1} \right)^{-1} \left( \frac{X^T y}{\sigma^2} + \Sigma_{\beta}^{-1} \beta_0 \right), \left(\frac{X^T X}{\sigma^2} + \Sigma_{\beta}^{-1} \right)^{-1} \right)
$$


**Q1.2:**

**Prior for $\sigma^2$:**
$$
\sigma^2 \sim \text{Inv-Gamma} (\alpha^*, \beta^*)
$$  

$$
p(\sigma^2) \propto (\sigma^2)^{-\alpha^* - 1} \exp \left(-\frac{\beta^*}{\sigma^2} \right)
$$  

**Likelihood:**
$$
y | X, \beta, \sigma^2 \sim N(X\beta, \sigma^2 I)
$$  

$$
p(y | X, \beta, \sigma^2) \propto (\sigma^2)^{-\frac{n}{2}} \exp \left( -\frac{1}{2\sigma^2} (y - X\beta)^T (y - X\beta) \right)
$$  

**Compute Posterior:**
$$
p(\sigma^2 | y, X, \beta) \propto p(y | X, \beta, \sigma^2) p(\sigma^2)
$$  

$$
= (\sigma^2)^{-\frac{n}{2}} \exp \left( -\frac{1}{2\sigma^2} S \right) \times (\sigma^2)^{-\alpha^* - 1} \exp \left( -\frac{\beta^*}{\sigma^2} \right)
$$  

where  

$$
S = (y - X\beta)^T (y - X\beta)
$$  

$$
= (\sigma^2)^{-(\alpha^* + \frac{n}{2} + 1)} \exp \left( -\frac{\beta^* + \frac{S}{2}}{\sigma^2} \right)
$$  

Thus,  
$$
\sigma^2 | y, X \sim \text{Inv-Gamma} \left( \alpha^* + \frac{n}{2}, \beta^* + \frac{1}{2} (y - X\beta)^T (y - X\beta) \right)
$$

**Answer:**

The following data analysis is performed using a dataset sourced from Kaggle. After cleaning the dataset (which involves renaming variables, handling missing observations, etc.), the cleaned dataset is loaded, containing various measurements on computers, such as price, weight, RAM, and size in inches.

This analysis aims to model the relationship between computer prices and their features. The regression model takes the form:

$$
\text{Price} \sim \beta_0 + \beta_1 \cdot \text{Inches} + \beta_2 \cdot \text{Ram} + \beta_3 \cdot \text{Weight} + \epsilon
$$

The following code first extracts and normalizes the numerical predictors. 

The Bayesian priors are defined as:

- $\beta_j \sim \mathcal{N}(0,1)$ for all coefficients.

- $\sigma \sim \text{HalfNormal}(1)$ (instead of an inverse-gamma prior).

Then, the model is fitted using PyMC, and MCMC sampling is performed.

```python
import pymc as pm
import numpy as np
import pandas as pd
import arviz as az
from patsy import dmatrix  # For creating design matrices

# Load the dataset 
df = pd.read_csv("cleaned_data (1).csv")  # Update with the actual filename
```

```python
# Selecting features and response of interest
features = ["Inches", "Ram", "Weight"]
target = "Price"

# Extract X (predictors) and y (response)
X = df[features].values
y = df[target].values

# Standardize features for better MCMC stability
X_mean, X_std = X.mean(axis=0), X.std(axis=0) + 1e-6  # Small epsilon to prevent div by zero
X = (X - X_mean) / X_std
y_mean, y_std = y.mean(), y.std()
y = (y - y_mean) / y_std  # Normalize target

n, p = X.shape  # Number of samples and predictors

# Bayesian Linear Regression in PyMC
with pm.Model() as model:
    # Priors for coefficients
    betas = pm.Normal('betas', mu=0, sigma=1, shape=p)  # Independent priors for coefficients
    sigma = pm.HalfNormal('sigma', sigma=1)  # Half-normal prior for sigma
    intercept = pm.Normal("intercept", mu=0, sigma=1)  # Prior for intercept

    # Linear model
    mu = intercept + pm.math.dot(X, betas)

    # Likelihood
    y_obs = pm.Normal('y', mu=mu, sigma=sigma, observed=y)

    # MCMC Sampling
    trace = pm.sample(2000, return_inferencedata=True, target_accept=0.9)
```

**Posterior Analysis & MCMC Diagnostics**

```python
# Posterior Analysis
az.summary(trace)
```

```python
import matplotlib.pyplot as plt
az.plot_trace(trace)
plt.tight_layout()
plt.show()
```

**Posterior analysis:**

The posterior analysis of the Bayesian linear regression model helps us identify the association between the predictors (Inches, Ram, and Weight) and the computer price. The posterior means indicate the average effect of each predictor on Price, while the highest density intervals (HDI) shows the uncertainty estimates. 

- Inches ($\hat{\beta_1}$ =0.076) has a small negative effect, with a credible interval [-0.128, -0.022]. Since the interval does not contain zero, there is moderate evidence that increasing screen size slightly reduces price (perhaps due to other correlated factors).

- Ram ($\hat{\beta_2}$ =0.648) has the strongest positive effect on Price, with a narrow 95% credible interval [0.600, 0.693], suggesting high certainty that increasing RAM leads to higher computer prices.

- Weight ($\hat{\beta_3}$ =0.018) has a very weak effect, with a credible interval [-0.035, 0.072] that overlaps zero, meaning the effect of computer's weight on their price is uncertain and might be negligible.

**MCMC Diagnostics:**

The R-hat values (all 1.0) confirm good convergence, meaning the sampled Markov chains mixed well.

The Effective Sample Size (ESS) is very high (all >6000), showing that the sampling process was efficient and reliable.

**Final Interpretation:**

The results suggest that RAM is the most significant factor affecting laptop price, with strong evidence of a positive effect. Screen size (measured in inches) has a small negative effect, while weight appears to have little to no impact. The model performs well with stable MCMC diagnostics, but some variance in price remains unexplained, as seen from the 
$\sigma$ estimate. Further improvements could involve adding more laptop features to capture more price variability.

**Question:**

Let $p(\tau_i)$ be $\require{cancel} \textrm{gamma}\big(\tau_i | \alpha = \frac{\nu}{2}, \overset{\textrm{rate}\xcancel{\textrm{scale}}}{\beta = \frac{\nu}{2}}\big)$ and let $p(y_i|\boldsymbol \beta, \tau,\tau_i)$ be $\mathcal{N}(y_i | \textbf{X} \boldsymbol \beta, \overset{\textrm{precision}}{\tau \times \tau_i})$. 

- Return to your dataset and maniputate it to have some various outliers or find another dataset with some outlier data
- Use an appropriate prior for inference on $v$ if you have enough data to do so
- Use the posterior distributions of the $\tau_i$'s to identify data point "outliers" 
- Use the posterior distributions of the $\sigma_i^{-2} = \tau \times \tau_i$ to create posterior distribuitions of the **influence** (the diagonals of the $H$ "hat" matrix $X^\top (X^\top D X)^{-1} X$ where $D_{ij}=0$ and $D_{ii} = \sigma^2_i$) and compare and contras some example "outlier" versus "non outlier" data points

- Provide inference with Bayesian posterior analysis and report MCMC diagnostics

**Define and fit the Bayesian Model with Varying Variance:**

```python
import pymc as pm
import arviz as az

# Extract predictors and response
features = ["Inches", "Ram", "Weight"]
X = df[features].values
y = df["Price"].values

# Standardize for stability
X_mean, X_std = X.mean(axis=0), X.std(axis=0) + 1e-6
X = (X - X_mean) / X_std
y_mean, y_std = y.mean(), y.std()
y = (y - y_mean) / y_std

n, p = X.shape  # Number of samples and predictors

# Bayesian Linear Regression with Varying Variance
with pm.Model() as model:
    # Priors for coefficients
    betas = pm.Normal('betas', mu=0, sigma=1, shape=p)
    intercept = pm.Normal("intercept", mu=0, sigma=1)

    # Studentized Variance Priors
    nu = pm.Gamma("nu", alpha=2, beta=0.1)  # Degrees of freedom
    tau = pm.Gamma("tau", alpha=2, beta=2)  # Global precision
    tau_i = pm.Gamma("tau_i", alpha=nu / 2, beta=nu / 2, shape=n)  # Individual precisions

    # Linear model
    mu = intercept + pm.math.dot(X, betas)

    # Likelihood with varying precision
    y_obs = pm.Normal('y', mu=mu, sigma=pm.math.sqrt(1 / (tau * tau_i)), observed=y)

    # MCMC Sampling
    trace = pm.sample(2000, return_inferencedata=True, target_accept=0.9)
```

```python
# Posterior Analysis
az.summary(trace)
```

```python
import matplotlib.pyplot as plt
az.plot_trace(trace)
plt.tight_layout()
plt.show()
```

**Identify outliers::**

```python
import matplotlib.pyplot as plt

# Extract tau_i values
tau_i_samples = trace.posterior["tau_i"].mean(dim=["chain", "draw"])

# Plot tau_i to identify potential outliers
plt.figure(figsize=(10, 4))
plt.scatter(range(n), tau_i_samples, color="red")
plt.axhline(np.percentile(tau_i_samples, 5), linestyle="--", color="black", label="5th Percentile")
plt.xlabel("Data Index")
plt.ylabel("Estimated Tau_i (Precision)")
plt.title("Identifying Outliers using Tau_i")
plt.legend()
plt.show()
```

**Visualize the influence of outliers:**

```python
import numpy as np
import matplotlib.pyplot as plt

# Convert tau_i_samples to 1D array
tau_i_samples = trace.posterior["tau_i"].mean(dim=["chain", "draw"]).values.flatten()

# Create diagonal matrix D
D_sqrt = np.diag(np.sqrt(tau_i_samples))  # Square root of D for proper weighting

# Apply weighting correctly
X_D = D_sqrt @ X  # Now X_D has shape (n, p)

# Compute influence (hat values)
H = np.diag(X_D @ np.linalg.inv(X_D.T @ X_D) @ X_D.T)

# Plot influence values
plt.figure(figsize=(10, 4))
plt.scatter(range(len(H)), H, color="blue")
plt.axhline(np.percentile(H, 95), linestyle="--", color="black", label="95th Percentile")
plt.xlabel("Data Index")
plt.ylabel("Influence (Hat Matrix Diagonal)")
plt.title("Identifying High Influence Points")
plt.legend()
plt.show()
```

This analysis applies a Bayesian hierarchical regression model to a dataset of laptop prices, using Inches, RAM, and Weight as predictors. Instead of artificially introducing outliers, the model is designed to detect naturally occurring anomalies in the data and account for heteroskedasticity. The variance structure ($\tau_i$) is incorporated, allowing each observation to have a unique variance rather than assuming constant error variance across all data points. The variance structure is modeled using Gamma priors on ($\tau_i$). After performing MCMC sampling, the posterior summary reveals a $\nu$ estimate of approximately 3.83, indicating heavy-tailed errors, which suggests the presence of extreme values or outliers in the dataset. The estimated regression coefficients ($\beta$) align with expectations, with RAM emerging as the strongest predictor of price, as evidenced by a credible interval excluding zero.

To further investigate outliers, the posterior distributions of $\tau_i$ are analyzed. A scatter plot of 
$\tau_i$ values highlights data points with low $\tau_i$, which correspond to high-variance observations. A 5th percentile threshold is used to classify outliers, confirming that some data points exhibit significantly higher uncertainty compared to the rest of the dataset. In addition, influence diagnostics are computed using the hat matrix diagonal (H), which quantifies the degree to which each observation affects regression predictions. The results indicate that certain high-influence points coincide with low $\tau_i$ values, demonstrating that outliers not only exhibit high variance but also exert strong leverage on the regression estimates. MCMC diagnostics, including trace plots and R-hat values, confirm that the model converges well, ensuring reliable posterior estimates.

The findings suggest that Bayesian regression effectively identifies naturally occurring outliers by incorporating adaptive variance through $\tau_i$. This approach distinguishes between high-variance and high-influence observations, providing a more robust alternative to traditional regression models that assume constant variance. By allowing for heteroskedasticity, the Bayesian framework offers greater flexibility in handling real-world data on laptop prices with anomalies.
